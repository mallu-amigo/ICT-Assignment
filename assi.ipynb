```python
# Convert the mpg column into a NumPy array and calculate: 
# ○ The mean, median, and standard deviation of mpg. 
# ○ The number of cars with mpg greater than 25. 

import numpy as np
import pandas as pd

data = pd.read_csv("auto-mpg.csv")
data_mpg = data['mpg'].to_numpy()

# mean of mpg 

my_mean = data_mpg.mean()
print(f"mean :{my_mean}")

# median of mpg

median = np.median(data_mpg)
print(f"median = {median}")


# finding the standard deviation

mean = np.mean(data_mpg) #finding the mean
sq_dev = (data_mpg - mean) ** 2
variance = np.sum(sq_dev) / (len(data_mpg) - 1)
standard_dev = np.sqrt(variance)
print(f"standard deviation = {standard_dev}")

print(data.columns.tolist())

```

```python
# Using NumPy, filter all cars with more than 6 cylinders. 
# Return the corresponding car_name as a list

car_name = data[data['cylinders'] > 6]['car name'].tolist()
print(car_name)
```

```python
# Compute the 25th, 50th, and 75th percentiles of the weight column using NumPy. 

weight = data['weight']
def percentile(x, p):
    return np.percentile(x, p)

percentile(weight, 25) #25 percentile
percentile(weight, 50) #50 percentile
percentile(weight, 75) #75 percentile
```

```python
# Convert the acceleration column into a NumPy array and normalize its values 
# (scale between 0 and 1). 

acceleration = data['acceleration'].to_numpy()
acc_norm = (acceleration - acceleration.min()) / (acceleration.max() - acceleration.min())
print(f"normalized values = {acc_norm}")
```

```python
# Increase all horsepower values by 10% and store the updated values in a new 
# NumPy array. Handle missing data (if any) by replacing it with the mean of the 
# column before applying the increase. 

horsepower = pd.to_numeric(data['horsepower'], errors='coerce')
mean_hp = np.nanmean(horsepower)
horsepower_filled = np.where(np.isnan(horsepower), mean_hp, horsepower)
updated_horsepower = horsepower_filled * 1.1
print(f"updated horsepower values = {updated_horsepower}")
```

```python

# Find the average displacement of cars with an origin of 2 (Europe) using NumPy indexing.


data['origin'] = data['origin'].replace({2: 'Europe'})
euro_cars = data[data['origin'] == 'Europe']['displacement']
avg_displacement = np.mean(euro_cars)
print(f"average displacement = {avg_displacement}")

```

```python
# Create a 2D NumPy array containing the columns mpg, horsepower, and weight. 
# Compute the dot product of this matrix with a given vector [1, 0.5, -0.2]. 

matrix = np.column_stack((data['mpg'], horsepower_filled, data['weight']))
vector = [1, 0.5, -0.2]

dot_product = np.dot(matrix, vector)
print(dot_product)




```

```python
# Use NumPy to sort the cars by model_year in descending order and display the first 
# f
#  ive car names.
sorted = np.argsort(-data['model year'].to_numpy())
first_five_names = data.iloc[sorted]['car name'].head(5).tolist()
print(first_five_names)

```

```python
# Compute the Pearson correlation coefficient between mpg and weight using 
mpg = data['mpg'].to_numpy()
weight_arr = data['weight'].to_numpy()
mpg_mean = np.mean(mpg)
weight_mean = np.mean(weight_arr)
numerator = np.sum((mpg - mpg_mean) * (weight_arr - weight_mean))
denominator = np.sqrt(np.sum((mpg - mpg_mean)**2) * np.sum((weight_arr - weight_mean)**2))
pearson_corr = numerator / denominator
print(f"Pearson correlation coefficient: {pearson_corr}")
```

```python
# Calculate the mean mpg for cars grouped by the number of cylinders using NumPy 
# techniques. 

mpg = data['mpg'].to_numpy()
cylinders = data['cylinders'].to_numpy()
unique_cylinders = np.unique(cylinders)

for cyl in unique_cylinders:
     mpg_for_cyl = mpg[cylinders == cyl] 
     mean_mpg = np.mean(mpg_for_cyl)
     print(f"Cylinders: {cyl}, Mean MPG: {mean_mpg:.2f}")
```

```python
# oad the dataset into a Pandas DataFrame. Display: 
# ○ The first 10 rows 
# ○ The total number of rows and columns 
# ○ Summary statistics for numerical columns

print(data.head(10))
print(f"Total rows: {data.shape[0]}, Total columns: {data.shape[1]}")
print(data.describe())
```

```python
# Find all cars manufactured in 1975 with a weight less than 3000. Return the 
# DataFrame with selected columns: car_name, weight, and mpg.

filtered_data = data[(data['model year'] == 75) & (data['weight'] < 3000)][['car name', 'weight', 'mpg']]
print(filtered_data)
```

```python
# Identify if there are any missing values in the dataset. Replace missing values in the 
# horsepower column with the column's median.


missing_values = data.isnull().sum()
print("Missing values per column:\n", missing_values)
hp_median = pd.to_numeric(data['horsepower'], errors='coerce').median()
data['horsepower'] = pd.to_numeric(data['horsepower'], errors='coerce').fillna(hp_median)
print(f"Median horsepower used = {hp_median}")
```

```python
# Add a new column power_to_weight_ratio, calculated as horsepower / weight.

data['power_to_weight_ratio'] = data['horsepower'] / data['weight']
print(data[['horsepower', 'weight', 'power_to_weight_ratio']].head())
```

```python

# Group the cars by origin and calculate the mean mpg for each group. 
cars = data['origin'].to_numpy()
unique_cars = np.unique(cars)

for car in unique_cars:
    mpg_for_origin = mpg[cars == car]
    mean_mpg = np.mean(mpg_for_origin)
    print(f"Origin: {car}, Mean MPG: {mean_mpg:.2f}")

```

```python
# Sort the DataFrame by mpg in descending order and display the top 10 cars with 
# the highest mpg. 

sorted = np.argsort(-data['mpg'].to_numpy())
def performance_score(row):
    return row['mpg'] * row['acceleration'] / row['weight']

data['performance_score'] = data.apply(performance_score, axis=1)
top10 = data.iloc[sorted][['car name', 'performance_score']].head(10)
print(top10)
```

```python
summary = data.groupby('model year')[['mpg', 'weight', 'horsepower']].mean().reset_index()
print(summary)
```

```python
# Save a subset of the data containing only mpg, cylinders, horsepower, and weight 
# for cars with mpg > 30 into a CSV file named high_mpg_cars.csv. 

subset = data[data['mpg'] > 30][['mpg', 'cylinders', 'horsepower', 'weight']]
subset.to_csv('high_mpg_cars.csv', index=False)
```

```python
# Identify potential outliers in the mpg column using the Interquartile Range (IQR) 
# method. Specifically: 
# ● Calculate the IQR for mpg. 
# ● Define outliers as values less than Q1 - 1.5 * IQR or greater than Q3 + 1.5 * 
# IQR. 
mpg = data['mpg'].to_numpy()
Q1 = np.percentile(mpg, 25)
Q3 = np.percentile(mpg, 75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = mpg[(mpg < lower_bound) | (mpg > upper_bound)]
print(f"IQR: {IQR}")
print(f"Lower bound: {lower_bound}, Upper bound: {upper_bound}")
print(f"Outliers in mpg: {outliers}")
```
